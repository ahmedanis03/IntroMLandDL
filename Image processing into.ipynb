# Image Processing Intro.

```python
import numpy as np
from skimage import data, filters
import matplotlib.pyplot as plt
from scipy import signal
%matplotlib inline
```

```python
img = data.coins()
plt.imshow(img,cmap='gray')
print(img.shape)
```

```python
print(img[:10,:10])
print(img.shape)
```

```python
edges = filters.sobel(img)
_,ax = plt.subplots(1,2,figsize=[12,12])
ax[0].imshow(img,cmap='gray')
ax[1].imshow(edges,cmap='gray')
```

```python
def sobel_filter(img):
     
    width, height = img.shape
 
    kh = np.array([[-1, 0, 1], [-2, 0, 2], [-1, 0, 1]], dtype = np.float)
    kv = np.array([[1, 2, 1], [0, 0, 0], [-1, -2, -1]], dtype = np.float)
    print(kh)
     
    gx = signal.convolve2d(img, kh, mode='same', boundary = 'symm', fillvalue=0)
    gy = signal.convolve2d(img, kv, mode='same', boundary = 'symm', fillvalue=0)
 
    g = np.sqrt(gx * gx + gy * gy)
    g *= 255.0 / np.max(g)
    
   
    return g

def identity_filter(img):
    width, height = img.shape
 
    filter = np.array([[0, 0, 0], [0, 1, 0], [0, 0, 0]], dtype = np.float)
    print(filter)
     
    applied = signal.convolve2d(img, filter, mode='same', boundary = 'symm', fillvalue=0)

    return applied
```

```python
identical_img = identity_filter(img)
_,ax = plt.subplots(1,2,figsize=[12,12])
ax[0].imshow(img,cmap='gray')
ax[1].imshow(identical_img,cmap='gray')
```

```python
edges = sobel_filter(img)
_,ax = plt.subplots(1,2,figsize=[12,12])
ax[0].imshow(img,cmap='gray')
ax[1].imshow(edges,cmap='gray')
```
