## How to train NN?

1. Feedforward phase
2. Calculate Loss

4. Go to Step 1

### Do we really need to calculate the gradient for all the data samples ?! 

```python
Image('NN_model_rep.png')
```

### stochastic  gradient decent:
1. Shuffle the data
2. take sample (mini batch)
3. calculate the gradient
4. update weights
5. go to data_finished (Epoch Finished)? go to step 1 : go to step 2

```python
Image('StocVsBatch.png')
```
